```python
import pandas as pd
import numpy as np
```

```python
## Importing the first csv
df1 = pd.read_csv("Sexual History Question.csv")


## Importing the second csv
df2 = pd.read_csv("Sexual History.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
merged_df = pd.merge(df2, df1, on='sexual_history_id', how='inner')
```

```python
merged_df = merged_df[['sexual_history_id', 'person_id', 'question', 'response_type']]

merged_df.head()
```

```python
#merged_df.drop_duplicates(subset=['sexual_history_id', 'person_id'], keep='first', inplace=True)
```

```python
# Pivot the DataFrame



merged_df = merged_df.pivot_table(index=['sexual_history_id', 'person_id'],columns='question', values='response_type' , aggfunc=lambda x: ' '.join(x))

merged_df = merged_df.reset_index()


merged_df.head()
```

```python
merged_df.to_csv('merged_csv.csv', index=False)
```

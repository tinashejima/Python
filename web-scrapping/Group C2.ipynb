```python
import nltk

nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
file= open('scraped_text.txt','r')
text = file.read()
print(text)
```

# Preprocess text

```python
# Sentence Segmantation
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)

sentences
```

```python
# Word Tokenization
from nltk.tokenize import word_tokenize
words = word_tokenize(text)

words
```

```python
# Lowercasing
words = [word.lower() for word in words]
words
```

```python
# Removing punctuation and numbers
from string import punctuation
words = [word for word in words if word not in punctuation and not word.isdigit()]
words
```

# Stopwords removal

```python
# Removing stop words
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
print(stop_words)
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = word_tokenize(text)

lemmatized_words = []
for word in words:
    lemmatized_words.append(lemmatizer.lemmatize(word))
word_lem = ' '.join(lemmatized_words)
```

```python
word_lem
```

```python
import spacy # library for NLP tasks
nlp = spacy.load("en_core_web_sm") # the variable that loads the smaller English model to work on the text in the article
nlp.pipe_names # to see what functions are contained within the loaded spacy's module
```

```python
article= nlp(word_lem) # storing the lemmatized words in the variable to do NER
```

```python
# to recognize entities in the article variable and the explanation for what each entity is
for ent in article.ents:
    print(ent.text,"||", ent.label_,"|", spacy.explain(ent.label_))
```

```python
#display the  full article, highlighting entities
from spacy import displacy

displacy.render(article, style="ent")
```

# why we chose Lemmatization:

Lemmatization considers the context and converts the word to its dictionary form (lemma) for example the word ideas was returned to idea. As a result, it produces cleaner and more accurate word forms. it is slower than stemming but better for understanding.
It is useful for tasks like Named Entity Recognition (NER) and text classification where preserving meaning and context is crucial.

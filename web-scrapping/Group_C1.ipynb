```python
file= open('scraped_text.txt','r')
text = file.read()
print(text)
```

# Preprocess text

```python
# Sentence Segmantation
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)

sentences
```

```python
# Word Tokenization
from nltk.tokenize import word_tokenize
words = word_tokenize(text)

words
```

```python
# Lowercasing
words = [word.lower() for word in words]
words
```

```python
# Removing punctuation and numbers
from string import punctuation
words = [word for word in words if word not in punctuation and not word.isdigit()]
words
```

# Stopwords removal

```python
# Removing stop words
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
print(stop_words)
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
stemmed_words = []
for word in words:
    stemmed_words.append(stemmer.stem(word))
word_stem = ' '.join(stemmed_words)
```

```python
word_stem
```

```python
for word in words:
    print(word, "|", stemmer.stem(word))
```

### Named Entity Recognition on the clean article after lemmatization

```python
import spacy # library for NLP tasks
nlp = spacy.load("en_core_web_sm") # the variable that loads the smaller English model to work on the text in the article
nlp.pipe_names # to see what functions are contained within the loaded spacy's module
```

```python
article= nlp(word_stem) # storing the stemmed words in the variable to do NER
```

```python
# to recognize entities in the article variable and the explanation for what each entity is
for ent in article.ents:
    print(ent.text,"||", ent.label_,"|", spacy.explain(ent.label_))
```

```python
#display the  full article, highlighting entities
from spacy import displacy

displacy.render(article, style="ent")
```

# ANSWER: WHY WE CHOSE STEMMING 

-Stemming is a simple method that cuts off the ends of words to reduce them to a “stem.” For example, “Zimbabwean” was trimmed to “zimbabw” and “activism” to “activ.” 

-While stemming was fast and easy, it produced word fragments that are not proper words. 

-This is problematic for NER because language models (like those in spaCy) are trained on complete, dictionary-standard words.

-When the tokens were stemmed, the system was not able to recognize names or organizations correctly for example, in our processed output, a truncated form might lose its meaning and not be detected as a NORP (for nationalities or political groups) or a GPE (for locations).

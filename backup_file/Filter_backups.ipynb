```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("./backup_stats_31102024.csv")  # Replace with your CSV file path

# Assuming 'submission_date' is in a valid date format (e.g., 'YYYY-MM-DD' or 'MM/DD/YYYY')
# Convert 'submission_date' to datetime
df['submission_date'] = pd.to_datetime(df['submission_date'], errors='coerce')

# Split 'submission_date' into separate 'day', 'month', and 'year' columns
df['day'] = df['submission_date'].dt.day
df['month'] = df['submission_date'].dt.month
df['year'] = df['submission_date'].dt.year

# Filter to keep only rows where the month is July (7), August (8), or September (9)
df_filtered = df[df['month'].isin([7, 8, 9])]

# # Further filter to keep only rows where the day is between 25 and 31
# df_filtered = df_filtered[(df_filtered['day'] >= 25) & (df_filtered['day'] <= 31)]


# Group by 'site_id', 'month', and 'year', and return the maximum 'day'
grouped_df = df_filtered.groupby(['site_id', 'month', 'year'], as_index=False).agg({'day': 'max'})

# Create new columns for July, August, and September based on month and submission_date range (25-31)
grouped_df['July'] = grouped_df.apply(lambda x: 1 if x['month'] == 7 and 25 <= x['day'] <= 31 else 0, axis=1)
grouped_df['August'] = grouped_df.apply(lambda x: 1 if x['month'] == 8 and 25 <= x['day'] <= 31 else 0, axis=1)
grouped_df['September'] = grouped_df.apply(lambda x: 1 if x['month'] == 9 and 25 <= x['day'] <= 31 else 0, axis=1)

# Select only the desired columns
output_df = grouped_df[['site_id', 'July', 'August', 'September']]
output_df
# Save the output to a new Excel file
output_file_path = 'f_backup.csv'  # Replace with your desired output file path
output_df.to_csv(output_file_path, index=False)

print("Processed data saved to:", output_file_path)
```

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("./backup_stats_31102024.csv")  # Replace with your CSV file path

# Print the first few rows of df
print(df.head())

# Convert 'submission_date' to datetime
df['submission_date'] = pd.to_datetime(df['submission_date'], errors='coerce')

# Split 'submission_date' into separate 'day', 'month', and 'year' columns
df['day'] = df['submission_date'].dt.day
df['month'] = df['submission_date'].dt.month
df['year'] = df['submission_date'].dt.year

# Filter to keep only rows where the month is July (7), August (8), or September (9)
df_filtered = df[df['month'].isin([7, 8, 9])]

# Further filter to keep only rows where the day is between 25 and 31
df_filtered = df_filtered[(df_filtered['day'] >= 25) & (df_filtered['day'] <= 31)]

# Group by 'site_id', 'month', and 'year', and return the maximum 'day'
grouped_df = df_filtered.groupby(['site_id', 'month', 'year'], as_index=False).agg({'day': 'max'})

# Create new columns for July, August, and September based on month and submission_date range (25-31)
grouped_df['July'] = grouped_df.apply(lambda x: 1 if x['month'] == 7 else 0, axis=1)
grouped_df['August'] = grouped_df.apply(lambda x: 1 if x['month'] == 8 else 0, axis=1)
grouped_df['September'] = grouped_df.apply(lambda x: 1 if x['month'] == 9 else 0, axis=1)

# Pivot the DataFrame to make sure all site_id rows are present, even if the values are zero
output_df = (
    grouped_df.pivot_table(index='site_id', values=['July', 'August', 'September'], aggfunc='max', fill_value=0)
    .reset_index()
)

# Save the output to a new CSV file
output_file_path = 'f_backup.csv'  # Replace with your desired output file path
output_df.to_csv(output_file_path, index=False)

print("Processed data saved to:", output_file_path)
output_df

print(df.shape)
```
